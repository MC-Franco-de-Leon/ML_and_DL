```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(newd[['RM','LSTAT','INDUS']],newd[['target']],random_state=123,test_size=.3)
```

```python
x_train.describe()
```

```python
x_test.describe()
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
import math
```

```python
y_pred=model.predict(x_test)
```

```python
print('rmse',math.sqrt(mean_squared_error(y_pred,y_test)))
```

```python
print('R^2 score',r2_score(y_pred,y_test))
```

```python
plt.scatter(x_train['LSTAT'],y_train,color='yellow')
plt.scatter(x_test['LSTAT'],y_test,color='red')
plt.scatter(x_test['LSTAT'],y_pred,color='blue')
plt.show()
```
