```python
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

```

```python
def GenerateDf():
    data=load_boston()
    X, Y=pd.DataFrame(data=data.data, columns=data.feature_names), pd.DataFrame(data=data.target, columns=['Target'])
    return train_test_split(X, Y, test_size=.3)
```

```python
x_train,x_test, y_train, y_test=GenerateDf()
```

```python
clf=tree.DecisionTreeRegressor(max_depth=3)
```

```python
model=clf.fit(x_train, y_train)
```

```python
tree.plot_tree(model)
plt.show()
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print('mse ', mean_squared_error(y_pred, y_test))
```

```python
plt.figure()
plt.plot(x_test[['LSTAT']], y_test,'ro')
plt.plot(x_test[['LSTAT']], y_pred, 'go')
plt.show()
```

```python
print('r2 score train',r2_score(y_train, model.predict(x_train)))
print('r2 score test', r2_score(y_test, y_pred))
```
